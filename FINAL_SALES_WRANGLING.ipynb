```python
cop
```

```python
cop['Total amount'] = grp2['Total']
for i in range(0,len(cop)):
    cop['Total amount'][i]= (grp2['Total'][i])
cop.to_csv("product inline.csv",index=False)
cop
```

```python
cop['Tax'] = grp2['Tax 5%']
for i in range(0,len(cop)):
    cop['Tax'][i]= (grp2['Tax 5%'][i])
cop.to_csv("product inline.csv",index=False)
cop
```
